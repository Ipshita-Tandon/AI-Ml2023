```python
import os
import numpy as np

# Define paths
files_path = '/content/free-spoken-gujarati-digit-dataset'
if not os.path.exists(files_path):
    os.makedirs(files_path)

x = []
y = []
total = 23500  # Total number of expected files (if known)
count = 0

# Main directory containing the subdirectories
main_dir = '/content/free-spoken-gujarati-digit-dataset'

# List all subdirectories (R1 - Central Zone, S1, etc.)
for region in os.listdir(main_dir):
    region_path = os.path.join(main_dir, region)
    if not os.path.isdir(region_path):
        continue

    # Process subdirectories like S1, S2, etc.
    for subfolder in os.listdir(region_path):
        subfolder_path = os.path.join(region_path, subfolder)
        if not os.path.isdir(subfolder_path):
            continue

        # Process each .wav file in the subfolder
        for file_name in os.listdir(subfolder_path):
            if not file_name.endswith('.wav'):  # Process only .wav files
                continue

            file_path = os.path.join(subfolder_path, file_name)
            print(f"Processing file: {file_path}")

            # Extract features from the .wav file
            x.append(CalculateMelSpectrogram(file_path))

            # Assign labels (modify logic as per your labeling criteria)
            label = subfolder  # Example: use subfolder name as label
            y.append(label)

            count += 1
            if count % 300 == 0:
                print(f"Remaining files: {total - count}")

# Convert lists to NumPy arrays after processing is complete
x = np.array(x)
y = np.array(y)

# Save the arrays to .npy files
np.save(os.path.join(files_path, "X_melspectrogram.npy"), x)
np.save(os.path.join(files_path, "y_melspectrogram.npy"), y)

print(f"Processing completed. Saved processed data in {files_path}.")
```

```python
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
def GetCNN():
  model = models.Sequential()

  model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 1)))
  model.add(layers.MaxPooling2D((2, 2)))
  model.add(layers.BatchNormalization())

  model.add(layers.Conv2D(64, (3, 3), activation='relu'))
  model.add(layers.MaxPooling2D((2, 2)))
  model.add(layers.BatchNormalization())

  model.add(layers.Conv2D(64, (3, 3), activation='relu'))
  model.add(layers.MaxPooling2D((2, 2)))
  model.add(layers.BatchNormalization())

  model.add(layers.Dropout(0.1))
  model.add(layers.Flatten())
  model.add(layers.Dense(256, activation='relu'))# fully connected
  model.add(layers.Dropout(0.1))
  model.add(layers.Dense(128, activation='relu'))
  model.add(layers.Dropout(0.1))

  model.add(layers.Dense(11,activation='softmax', use_bias=True))
  return model
```

```python
from tensorflow.keras.utils import to_categorical

# Adjust num_classes based on the dataset
num_classes = max(max(y_train), max(y_test)) + 1# Dynamically calculate number of classes
print("Number of classes:", num_classes)

# One-hot encode the labels
train_y = to_categorical(y_train, num_classes)
test_y = to_categorical(y_test, num_classes)

# Verify the shapes
print("train_y shape:", train_y.shape)
print("test_y shape:", test_y.shape)
```

```python
save_path = '/content/drive/MyDrive/auddsavedfiles/models'


from tensorflow.keras.callbacks import ModelCheckpoint
cnn_model = GetCNN()
print(cnn_model.summary())
cnn_model.compile(loss  = tf.keras.losses.CategoricalCrossentropy(),
              metrics = ["accuracy"],
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.01))
callbacks = ModelCheckpoint(save_path+"/melspectrogram_model_.keras", monitor='val_accuracy',
                                    mode='max',
                                    save_best_only=True,
                                    verbose=1)

print("X train shape = ",x_train.shape)
print("y train shape = ",train_y.shape)

history = cnn_model.fit(x_train, train_y, batch_size=64, epochs=150, verbose=1, validation_split=0.1, callbacks=[callbacks])
#import keras
#cnn_model = keras.models.load_model(save_path+"/melspectrogram_model_.h5")
#print("Test Accuracy is ", cnn_model.evaluate(x_test,test_y)[1])
```

```python
# Model Ensembling
from keras.models import Model
from keras import layers

def ensembleModels(models, model_input):
    # collect outputs of models in a list
    model_outputs=[model(model_input) for model in models]
    # averaging outputs
    ensemble_output=layers.average(model_outputs)
    # build model from same input and avg output
    modelEns = Model(inputs=model_input, outputs=ensemble_output,    name='ensemble')

    return modelEns
```

```python
from keras.layers import Input  # Correct import for Input
from keras.models import Model  # Ensure you have this for model creation

# Assuming ensembleModels is defined elsewhere
models = [mp_model, mfcc_model, zc_model]

model_input = Input(shape=models[0].input_shape[1:])  # c*h*w
modelEns = ensembleModels(models, model_input)  # Assuming this function is implemented
modelEns.summary()

modelEns.save(save_path + "/ensemble_.keras", include_optimizer=True)
```
